```python
import os


if 'tfg' not in os.listdir():
    os.chdir('..')
```

```python
from tfg.alphaZero import create_alphazero
from tfg.alphaZeroCallbacks import Checkpoint
from tfg.alphaZeroAdapters import ConnectNAdapter
from tfg.util import enable_gpu
from tfg.alphaZeroConfig import AlphaZeroConfig
from game.connect_n import ConnectN

import time
```

```python
enable_gpu()

game = ConnectN()
```

```python
# This training ended with an error after 1320 games and about 14 hours
config = AlphaZeroConfig(
    learning_rate=.01,
    regularizer_constant=.0001,
    momentum=.0,
    residual_layers=12,
    filters=128,
    kernel_size=(3, 3)
)

checkpoint = Checkpoint(directory='models/checkpoints/connect4', delay=2)

start = time.time()

alphazero = create_alphazero(game, ConnectNAdapter(game), max_workers=12,
                             self_play_times=120, max_games_counter=2400,
                             buffer_size=6000, batch_size=512, temperature=20,
                             callbacks=[checkpoint], epochs=20,
                             c_puct=1, exploration_noise=(.25, .06),
                             mcts_iter=600, nn_config=config)

print("Finished training after", time.time() - start, "seconds")
```
