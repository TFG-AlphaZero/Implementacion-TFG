```python
from game.tictactoe import TicTacToe
from game.connect_n import ConnectN

from tfg.strategies import HumanStrategy
from tfg.alphaZero import AlphaZero
from tfg.alphaZeroAdapters import TicTacToeAdapter, ConnectNAdapter
from tfg.util import play, enable_gpu
```

```python
enable_gpu()
```

```python
game = TicTacToe()
adapter = TicTacToeAdapter()

human = HumanStrategy(game, render_mode='plt')
alphazero = AlphaZero(game, adapter, mcts_iter=200)
alphazero.load('models/TicTacToe.h5')
```

```python
play(game, human, alphazero, render='plt', print_results=True)
```

```python
game = ConnectN()
adapter = ConnectNAdapter(game)

human = HumanStrategy(game, render_mode='plt')
alphazero = AlphaZero(game, adapter, mcts_iter=1200)
alphazero.load('models/Connect4.h5')
```
